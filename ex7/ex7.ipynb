Part 1: Find Closest Centroids

```python
def findClosestCentroids(X: np.array, centroids: np.array) -> np.array:
    m, d = np.shape(X)
    K, d = np.shape(centroids)
    result = np.zeros(m)
    distance = np.zeros((m,K))
    for j in range(K):
        centroid = centroids[j,:]
        X_centroid = X - centroid
        distance[:,j] = np.sum(X_centroid*X_centroid,axis=1)
    return np.argmin(distance,axis=1)
```

```python
# Select an initial set of centroids
K = 3 # 3 Centroids
initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])
idx = findClosestCentroids(X, initial_centroids);

print('Closest centroids for the first 3 examples: \n')
print('{}'.format(idx[:3]))
print('(the closest centroids should be 0, 2, 1 respectively)\n') #Different from octave as index starts at 0 and not 1
```

Part 2: Compute Means

```python
def computeCentroids(X: np.array, idx: np.array, K: int) -> np.array:
    m, n = np.shape(X)
    result = np.zeros((K,n))
    for i in range(K):
        is_i_centroid = np.array([np.int32(idx == i)*n]).T
        result[i,:] = np.sum(X*is_i_centroid,axis=0)/np.sum(is_i_centroid)
    return result
```

```python
centroids = computeCentroids(X, idx, K)
print('Centroids computed after initial finding of closest centroids: \n')
np.set_printoptions(precision=6, suppress=True)
print(centroids)
print('\n(the centroids should be\n')
print('[ 2.428301 3.157924 ]')
print('[ 5.813503 2.633656 ]')
print('[ 7.119387 3.616684 ]')
```

Part 3: K-Means Clustering

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm

import numpy as np
    
def plotDataPoints(X: np.array, idx: np.array, K: int)-> None:
    colors = cm.rainbow(np.linspace(0, 1, K))
    for i in range(K):
        centroidData = X[idx==i,:]
        plotColor = colors[i]
        plotColor[3] = 0.1 # For lighted color
        plt.scatter(centroidData[:,0],centroidData[:,1],color=plotColor)

def plotProgresskMeans(X: np.array, idx: np.array, previous_centroids: np.array) -> None:
    maxIter, K, n = np.shape(previous_centroids)
    plotDataPoints(X, idx, K)
    colors = cm.rainbow(np.linspace(0, 1, K))
    for k in range(K):
        plotColor = colors[k]
        for it in range(max_iters-1):
            prevC = previous_centroids[it,k]
            nextC = previous_centroids[it+1,k]
            plotColor[3] = (it+1.0)/(max_iters-1.0)
            plt.plot(prevC[0],prevC[1],color=plotColor,marker='x',markersize=20)
            plt.arrow(prevC[0],prevC[1],nextC[0]-prevC[0],nextC[1]-prevC[1],head_width=0.2,head_length=.2)

def runkMeans(X: np.array, initial_centroids: np.array, max_iters: int, plotProgression: bool = False) -> (np.array, np.array):
    m, n = np.shape(X)
    K, n = np.shape(initial_centroids)
    centroids = initial_centroids
    previous_centroids = np.array([centroids]) #We will store the history of the centroids to plot the move
    idx = np.zeros(m)
    for i in range(max_iters):
        idx = findClosestCentroids(X=X, centroids=centroids)
        centroids = computeCentroids(X=X, idx=idx, K=K)
        if plotProgression:
            previous_centroids = np.r_['0',previous_centroids,np.array([centroids])]
    if plotProgression:
        plotProgresskMeans(X, idx, previous_centroids)
    return centroids, idx
```

Part 4: K-Means Clustering on Pixels

```python
import scipy.ndimage 
A = scipy.ndimage.imread('bird_small.png')/255
imageSize = A.shape
X = A.reshape((imageSize[0]*imageSize[1],3)) #Each row is a pixel of the image
```

```python
def kMeansInitCentroids(X: np.array, K: int) -> np.array:
    m,n = np.shape(X)
    return X[np.random.permutation(m)[:K],:]
```

```python
K = 16
max_iters = 10

initial_centroids = kMeansInitCentroids(X, K)
centroids, idx = runkMeans(X, initial_centroids, max_iters)
```
